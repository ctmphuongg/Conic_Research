```sage
import re
import matplotlib.pyplot as plt
data = []

with open("experiment1.txt", "r") as f:
    content = f.read()

tests = content.strip().split("Test ")[1:]

for test in tests:
    n_match = re.search(r"N\s*=\s*(\d+)\s*=\s*(\d+)\s*\*\s*(\d+)", test)
    if not n_match:
        continue
    N = int(n_match.group(1))
    p = int(n_match.group(2))
    q = int(n_match.group(3))

    B_match = re.search(r"B\s*=\s*(\d+)", test)
    B = int(B_match.group(1)) if B_match else None
    
    pollard_match = re.search(r"Testing Pollard's method\s*\n\s*(Success|Failed)", test)
    pollard_result = pollard_match.group(1) if pollard_match else "Unknown"
    williams_match = re.search(r"Testing Williams' method\s*\n\s*(Success|Failed)", test)
    williams_result = williams_match.group(1) if williams_match else "Unknown"
    pell_match = re.search(r"Testing Pell's method\s*\n\s*(Success|Failed)", test)
    pell_result = pell_match.group(1) if pell_match else "Unknown"
    
    data.append({"N": N, "p": p, "q": q, "B": B, "pollard": pollard_result, "williams": williams_result, "pell": pell_result})
```

```sage
import ast
from sympy import factorint

def largest_prime_factor(n):
    factors = factorint(n)  # returns {prime: exponent}
    return max(factors.keys())

lp_p_minus_1 = []
lp_q_minus_1 = []
lp_p_plus_1  = []
lp_q_plus_1  = []
B_values     = []

for entry in pollard_failures:
    p = entry['p']
    q = entry['q']
    B = entry['B']
    
    lp_p_minus_1.append(largest_prime_factor(p - 1))
    lp_q_minus_1.append(largest_prime_factor(q - 1))
    lp_p_plus_1.append(largest_prime_factor(p + 1))
    lp_q_plus_1.append(largest_prime_factor(q + 1))
    B_values.append(B)

x = range(len(pollard_failures))
```

```sage
R_W = QuadraticField(d, 'sqrtd')
G = R_W.automorphisms()
t = a + b*R_W.0
tbar = G[1](t) 

x = tbar / t
assert x * G[1](x) == 1

ZZN = IntegerModRing(N)
z = polygen(ZZN, 'z')
x_williams = (ZZN.extension(z^2 - d, 'a')(x))
```

```sage
from sage.all import gcd, inverse_mod, Integers, var, prime_range
from random import randint

from sage.all import Integers, cached_function

def add_point(P1, P2, d, R):
    """
    Add two points on conic
    """
    r, s = P1
    t, u = P2
    x3 = r*t + s*u*d
    y3 = r*u + s*t
    return R(x3), R(y3)

def self_add_optimized(n, P, delta, R):
    """
    Optimized addition using 2 power  
    """
    if n == 0:
        return R(1), R(0)

    result = (R(1), R(0)) 
    first = True
    for bit in bin(n)[2:]: 
        if not first:
            result = add_point(result, result, delta, R)  # double

        if bit == '1':
            if first: # First bit, result = P
                result = P
            else: # After that, double prev result, add another P if bit = 1
                result = add_point(result, P, delta, R)  # add P
        
        first = False  # after first bit processed
    return result
```

```sage
prime_axis = []
pell_x_coords = []
pell_y_coords = []
williams_u_coords = []
williams_v_coords = []
# differences = []

for l in primes(2, B+1):
    e = ceil(log(N, l))
    E = l^e
    
    # Update Pell-Conic
    x_pell, y_pell = self_add_optimized(E, (x_pell, y_pell), d, R)
    
    # Update Williams
    x_williams = x_williams^E
    u, v = x_williams.list()
    
    prime_axis.append(l)
    pell_x_coords.append(int(x_pell))
    pell_y_coords.append(int(y_pell))
    williams_u_coords.append(int(u))
    williams_v_coords.append(int(v))
    # differences.append(abs(int(x_pell) - int(u)))

    print(f"{l:<6} | ({x_pell}, {y_pell}) | ({u}, {v})")
    if x_pell == 1 and y_pell == 0 and u == 1 and v == 0:
        break
```

```sage
import matplotlib.pyplot as plt

num_points = min(10, len(pell_x_coords))

x_pell_10 = pell_x_coords[:num_points]
y_pell_10 = pell_y_coords[:num_points]

u_10 = williams_u_coords[:num_points]
v_10 = williams_v_coords[:num_points]

plt.figure(figsize=(8, 8))
plt.scatter(x_pell_10, y_pell_10,
            marker='^',
            s=100,
            label="Pell (x_pell, y_pell)")
plt.scatter(u_10, v_10,
            marker='o',
            s=100,
            label="Williams (u, v)")

# Connect points in order to show trajectory
plt.plot(x_pell_10, y_pell_10, linestyle='--', alpha=0.5)
plt.plot(u_10, v_10, linestyle='--', alpha=0.5)

plt.xlabel("x_pell / u")
plt.ylabel("y_pell / v")
plt.title("First 10 Iterations: Pell vs Williams Trajectories")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
